```python
import sys
sys.path.insert(0, '..')
from dependencies import *
```

```python
# DATA PROCESSING OF ENVELOPES

# Resample spectrograms to 100 hz, offset and filter (data processing)
# Log
envelopes_log = [eelbrain.load.unpickle(PREDICTOR_DIR / f'{stimulus}~envelope-log.pickle') for stimulus in STIMULI]
envelopes_log = [x.bin(0.01, dim='time', label='start') for x in envelopes_log]
envelopes_log = [eelbrain.pad(x, tstart=-0.100, tstop=x.time.tstop + 1, name='envelope_log') for x in envelopes_log]
envelopes_log = [eelbrain.filter_data(x, LOW_FREQUENCY, HIGH_FREQUENCY) for x in envelopes_log]
# Log 8 band
envelopes_log_8band = [eelbrain.load.unpickle(PREDICTOR_DIR / f'{stimulus}~envelope-log-8band.pickle') for stimulus in STIMULI]
envelopes_log_8band = [x.bin(0.01, dim='time', label='start') for x in envelopes_log_8band]
envelopes_log_8band = [eelbrain.pad(x, tstart=-0.100, tstop=x.time.tstop + 1, name='envelope_log_8band') for x in envelopes_log_8band]
envelopes_log_8band = [eelbrain.filter_data(x, LOW_FREQUENCY, HIGH_FREQUENCY) for x in envelopes_log_8band]
# Onset
envelopes_onset = [eelbrain.load.unpickle(PREDICTOR_DIR / f'{stimulus}~envelope-onset.pickle') for stimulus in STIMULI]
envelopes_onset = [x.bin(0.01, dim='time', label='start') for x in envelopes_onset]
envelopes_onset = [eelbrain.pad(x, tstart=-0.100, tstop=x.time.tstop + 1, name='envelope_onset') for x in envelopes_onset]
envelopes_onset = [eelbrain.filter_data(x, LOW_FREQUENCY, HIGH_FREQUENCY) for x in envelopes_onset]
# Onset 8 band
envelopes_onset_8band = [eelbrain.load.unpickle(PREDICTOR_DIR / f'{stimulus}~envelope-onset-8band.pickle') for stimulus in STIMULI]
envelopes_onset_8band = [x.bin(0.01, dim='time', label='start') for x in envelopes_onset_8band]
envelopes_onset_8band = [eelbrain.set_time(x, gt.time, name='envelope_onset_8band') for x, gt in zip(envelopes_onset_8band, envelopes_log)]
envelopes_onset_8band = [eelbrain.filter_data(x, LOW_FREQUENCY, HIGH_FREQUENCY) for x in envelopes_onset_8band]
# Words
envelopes_words = [eelbrain.load.unpickle(PREDICTOR_DIR / f'{stimulus}~envelope-words.pickle') for stimulus in STIMULI]
envelopes_words_onset = [eelbrain.event_impulse_predictor(gt.time, data=data, name='word') 
                         for gt, data in zip(envelopes_log, envelopes_words)]
envelopes_words_lexical = [eelbrain.event_impulse_predictor(gt.time, value='lexical', data=data, name='lexical')
                          for gt, data in zip(envelopes_log, envelopes_words)]
envelopes_words_nlexical = [eelbrain.event_impulse_predictor(gt.time, value='nlexical', data=data, name='non_lexical') 
                           for gt, data in zip(envelopes_log, envelopes_words)]
```

```python
# ————————————————————————————————————————————————————————————————————————————————————————————————
# Save pickles

eelbrain.save.pickle(envelopes_log, 
                        PROCESSED_PREDICTOR_DIR / f'~processed_envelopes-log.pickle') # <-- This is referred to as gammatone-1 in document
eelbrain.save.pickle(envelopes_onset, 
                        PROCESSED_PREDICTOR_DIR / f'~processed_envelopes-onset.pickle') # <-- This is referred to as gammatone-on-1 in document
eelbrain.save.pickle(envelopes_log_8band, 
                        PROCESSED_PREDICTOR_DIR / f'~processed_envelopes-log-8band.pickle') # <-- This is referred to as gammatone-8 in document
eelbrain.save.pickle(envelopes_onset_8band, 
                        PROCESSED_PREDICTOR_DIR / f'~processed_envelopes-onset-8band.pickle') # <-- This is referred to as gammatone-on-8 in document
eelbrain.save.pickle(envelopes_words_onset, 
                        PROCESSED_PREDICTOR_DIR / f'~processed_envelopes-words-onset.pickle')
eelbrain.save.pickle(envelopes_words_nlexical, 
                        PROCESSED_PREDICTOR_DIR / f'~processed_envelopes-words-nlexical.pickle')
eelbrain.save.pickle(envelopes_words_lexical, 
                        PROCESSED_PREDICTOR_DIR / f'~processed_envelopes-words-lexical.pickle')

print(envelopes_log)
```
