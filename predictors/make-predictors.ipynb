```python
import sys
sys.path.insert(0, '..')
from dependencies import *
```

```python
# CREATE PREDICTORS (gammatone envelopes): envelope_log, envelope_onset, and envelope_words

# Word-to-word predictor variables for envelopes_words
word_table = eelbrain.load.tsv(STIMULUS_DIR / 'AliceChapterOne-EEG.csv')
word_table['InvLogFreq'] = 17 - word_table['LogFreq']

for i in STIMULI:
    # Standarize iterator to integer
    i = int(i)

    # ————————————————————————————————————————————————————————————————————————————————————————————————
    # Define all output paths
    paths = [
        PREDICTOR_DIR / f'{i}~envelope-log.pickle',
        PREDICTOR_DIR / f'{i}~envelope-onset.pickle',
        PREDICTOR_DIR / f'{i}~envelope-log-8band.pickle',
        PREDICTOR_DIR / f'{i}~envelope-onset-8band.pickle',
        PREDICTOR_DIR / f'{i}~envelope-words.pickle',
    ]
    # Check if all files exist AND are loadable
    for path in paths:
        skip = False
        if path.exists():
            try:
                existing_data = eelbrain.load.unpickle(path)
                if existing_data is not None:
                    skip = True
            except Exception as e:
                print(f"Warning: existing predictor file {path.name} is corrupted or unreadable, will recompute. ({e})")
    if skip:
        print(f"All predictors for stimulus {i} already exist and are valid, skipping...")
        continue
    print(f"Processing stimulus {i}")

    # ————————————————————————————————————————————————————————————————————————————————————————————————
    # Load the sound file corresponding to the predictors
    wav = eelbrain.load.wav(STIMULUS_DIR / f'{i}.wav')
    # Apply a filter to make it high resolution
    envelope = eelbrain.gammatone_bank(wav, 80, 15000, 128, location='left', tstep=0.001) # <-- Don't know yet how this works

    # ————————————————————————————————————————————————————————————————————————————————————————————————
    # Make the log envelope, and 8 band frequency version
    envelope_log = (envelope + 1).log()
    envelope_log_sum = envelope_log.sum('frequency')
    envelope_log_8band = envelope_log.bin(nbins=8, func='sum', dim='frequency')
    
    # Make the onset envelope, and 8 band frequency version
    envelope_onset = eelbrain.edge_detector(envelope_log, c=30)
    envelope_onset_sum = envelope_onset.sum('frequency')
    envelope_onset_8band = envelope_onset.bin(nbins=8, func='sum', dim='frequency')
    
    # Make the words envelope, and 8 band frequency version
    segment_table = word_table.sub(f"Segment == {i}")
    word_data = eelbrain.Dataset({'time': segment_table['onset']}, info={'tstop': segment_table[-1, 'offset']})
    word_data['LogFreq'] = segment_table['InvLogFreq']
    for key in ['NGRAM', 'RNN', 'CFG', 'Position']:
        word_data[key] = segment_table[key]
    word_data['lexical'] = segment_table['IsLexical'] == True
    word_data['nlexical'] = segment_table['IsLexical'] == False
    
    # ————————————————————————————————————————————————————————————————————————————————————————————————
    # Save pickles
    eelbrain.save.pickle(envelope_log_sum, 
                         PREDICTOR_DIR / f'{i}~envelope-log.pickle') # <-- This is referred to as gammatone-1 in document
    eelbrain.save.pickle(envelope_onset_sum, 
                         PREDICTOR_DIR / f'{i}~envelope-onset.pickle') # <-- This is referred to as gammatone-on-1 in document
    eelbrain.save.pickle(envelope_log_8band, 
                         PREDICTOR_DIR / f'{i}~envelope-log-8band.pickle') # <-- This is referred to as gammatone-8 in document
    eelbrain.save.pickle(envelope_onset_8band, 
                         PREDICTOR_DIR / f'{i}~envelope-onset-8band.pickle') # <-- This is referred to as gammatone-on-8 in document
    eelbrain.save.pickle(word_data, 
                         PREDICTOR_DIR / f'{i}~envelope-words.pickle') # <-- This is referred to as word_tables in document
print('Process Finished')
```
